# IMPORT LIBRARIES

```python
import numpy as np
from numpy import unique, argmax
from tensorflow.keras.datasets.mnist import load_data 
from tensorflow.keras import Sequential  
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout
from tensorflow.keras.utils import plot_model
import matplotlib.pyplot as plt
```

# LOADING MNIST DATASET

```python
(x_train ,y_train),(x_test,y_test)=load_data()
```

# RESHAPING THE TRAINING AND TESTING DATA 


```python
x_train = x_train.reshape((x_train.shape[0], x_train.shape[1], x_train.shape[2], 1))
x_test = x_test.reshape((x_test.shape[0], x_test.shape[1], x_test.shape[2], 1))
```

# NORMALIZING THE VALUES OF PIXELS OF IMAGES

```python
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
```

```python
fig = plt.figure(figsize=(5, 3))
for i in range(15):
    ax = fig.add_subplot(2, 10, i + 1, xticks=[], yticks=[])
    ax.imshow(x_train[i], cmap='gray') 
    ax.set_title(y_train[i])
```

# DETERMINE THE SHAPE OF INPUT IMAGES

```python
img_shape = x_train.shape[1:] 
print(img_shape)
```

# DEFINING THE MODEL

```python
model = Sequential()

# First Conv2D Layer
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=img_shape))

# MaxPooling Layer
model.add(MaxPool2D((2, 2)))

# Second Conv2D Layer
model.add(Conv2D(48, (3, 3), activation='relu'))

# MaxPooling Layer
model.add(MaxPool2D((2, 2)))

# Dropout Layer
model.add(Dropout(0.5))

# Flatten Layer
model.add(Flatten())

# Fully Connected Layer
model.add(Dense(500, activation='relu'))

# Output Layer
model.add(Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
plot_model(model, 'test_model.jpg', show_shapes=True)


```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
x=model.fit(x_train,y_train,epochs=10,batch_size=128,verbose=2,validation_split=0.1)
```

```python
# Evaluate the model on the test dataset
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)

# Print the accuracy as a percentage
print(f'Accuracy: {accuracy * 100:.2f}%')

```

```python
image = image.reshape(1, image.shape[0], image.shape[1], image.shape[2])  # (1, 28, 28, 1)

probabilities = model.predict(image)
print("Class probabilities:", probabilities)
print("Predicted class:", argmax(probabilities))
```
